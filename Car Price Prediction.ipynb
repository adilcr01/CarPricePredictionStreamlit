```python
cat_col=[]
int_col=[]
for i in df:
    if df[i].dtypes=='object':
        cat_col.append(i)
    else:
        int_col.append(i)
```

```python
cat=df[cat_col]
```

```python
cat
```

```python
integer_col=pd.DataFrame(sc.fit_transform(integer_col),columns=integer_col.columns)
integer_col
```

```python
from sklearn.metrics import r2_score,mean_squared_error
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from xgboost import XGBRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.ensemble import AdaBoostRegressor
```

```python
rfg=RandomForestRegressor(random_state=0)
lr=LinearRegression()
dtg=DecisionTreeRegressor(random_state=0)
xgbr=XGBRegressor()
gbr=GradientBoostingRegressor()
abr=AdaBoostRegressor()
```

```python
models={
    'Randomforest':rfg,
    'LinearRegression':lr,
    'DecisionTree':dtg,
    'XGboost':xgbr,
    'GradientBoostingRegressor':gbr,
    'AdaBoostRegressor':abr
}
```

```python
#Create a function to train multiple model at once.
model_list=[]
train_r2_list=[]
r2_squared_list=[]
def train(model,X_train,X_test,y_train,y_test):
    for model_name,model_obj in model.items():
        model_obj.fit(X_train,y_train)
        y_pred=model_obj.predict(X_test)
        train_r2_list.append(model_obj.score(X_train,y_train))
        r2_squared_list.append(r2_score(y_test,y_pred))
        model_list.append(model_name)
    return model_list,train_r2_list,r2_squared_list
```

```python
train(models,X_train,X_test,y_train,y_test)
```

```python
performance=pd.DataFrame([model_list,train_r2_list,r2_squared_list]).T
performance.rename(columns={0:'Algorithm',1:'Train_data_r2',2:'r2_model'},inplace=True)
```
